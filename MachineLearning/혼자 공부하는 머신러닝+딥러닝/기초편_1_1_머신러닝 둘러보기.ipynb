### 머신러닝 둘러보기 1. 용어

해당 자료에서는 도미가 두개 있으므로 2개의 클래스를 사용한다.

우리가 사용할 데이터는 도미와 빙어 두마리를 구분하고 분류할 것이기 때문에 이진분류(binary classification)이라고 말할 수 있다.

### 데이터
위 데이터에서는 생선의 '길이'와 '무게' 두개의 데이터를 가지고 있다.  
  
머신러닝에서는 각각의 데이터를 **샘플**이라고 한다. 따라서 위 그림에서는 35개의 샘플이 있다고 설명할 수 있다.  
또한 이 샘플의 길이, 무게 각각의 원소는 **특성**이라고 한다. 
**샘플**과 **특성**은 해당 요소를 지칭하는 가장 대표적인 명칭이라고 할 수 있다.(아님말고)  
각종 서적에 따라 부르는 방식이 조금씩 다를 수는 있지만 혹시 다른 명칭으로 머리가 복잡해지더라도 정신건강을 위해 그저 용어의 차이라고만 생각해두자  



````python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7,
                31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5,
                34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0,
                38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0,
                450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0,
                700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 
                925.0, 975.0, 950.0]
````

### 산점도(scatter plot)
앞서 본 그림처럼 특성이 두개 이상일 때,
데이터를 가장 잘 표현할 수 있는 방식 중 하나는 산점도(scatter plot)이다.  
x축과 y축 두개의 특성을 두고 각 샘플을 하나의 점으로 표현하는 것을 말한다.  
위 그림에서는 길이 특성을 x축에 두었고 무게 특성을 y축에 두었다. 사실 바뀌어도 상관없다. 그리는 사람 마음이다.  
파이썬에서 그래프를 그리는 가장 대표적인 라이브러리가 위에서 사용된 'matplotlib' 이다.

image.png

````python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

우리가 알고 있는 일반 상식처럼 크기가 증가하면 무게가 증가하는 것을 확인할 수 있다.

image.png

다음은 빙어 데이터이다.


````python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2,
                12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4,
                12.2, 19.7, 19.9]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

scatter() 함수를 두번 연속으로 호출하여 하나의 그래프에 두개의 데이터를 같이 중첩해서 볼 수 있다.  
(도미 데이터(파란색)와 빙어 데이터(붉은색)으로 그려진 것을 확인할 수 있다.)  
해당 데이터를 봐서는 크기와 무게로 두 물고기를 분류하는 것이 어렵지 않아보인다.

이제 머신러닝을 학습시키기 위해서는 해당 데이터를 하나로 만들어줘야한다.  
즉, 도미 데이터와 빙어 데이터를 하나로 합쳐 데이터의 정보를 학습함으로서 도미와 빙어를 구분할 수 있는 능력을 키워주는 것이라고 생각하면 된다.(똑같은 말인가?)

image.png

````python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight 
````

이제 여기서 끝나는 것이 아니라 데이터의 변경이 필요하다.  
그림 오른쪽처럼 샘플 하나의 데이터가 하나의 리스트에 담기도록, 또한 외곽쪽에 리스트가 있도록 만들어줘야한다.  
이유는 사이킷런 라이브러리가 앞서 말한 방식의 데이터(**2차원 배열** 또는 리스트의 리스트)를 요구하기 때문이다.  


image.png

````python
fish_data = [[l, w] for l, w in zip(length, weight)]
fish_data
````

````python
fish_target = [1]*35 + [0]*14
fish_target
````

<br/><br/>
# k-최근접 이웃 모델 이용하기

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target) # 모델(다른 클래스도 다 똑같이 fit매서드를 사용한다.)
kn.score(fish_data, fish_target)
````

# 새로운 생선 예측

길이 30cm, 무게 600g의 새로운 생선이 있다고 가정하자.  
이 생선의 종류는 무엇으로 분류될 수 있을까?  
해당 가상의 데이터가 가지고 있는 특징을 가지고 우리는 도미라는 것을 알 수 있다(길이와 무게를 통해)  
이 때 예측을 위해 사용하는 함수가 predict이다.


````python
kn.predict([[30, 600]])
````

아직까지 알고리즘이라고 하기엔 부족하지만, 위와 같은 방식으로 규칙을 찾고 정답을 예측한다는 것만 기억해두자.
<br/><br/>

# 무조건 도미

````python
kn49 = KNeighborsClassifier(n_neighbors=49) #n_neighbors : 클래스의 매개변수를 지정

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
````

현재 우리가 가진 생선 데이터의 갯수는 49개이다.  
따라서 ```n_neighbors```를 49로 지정할 경우 전체 샘플을 다 바라보게(?)된다. 
현재 전체 샘플의 다수는 도미이다. 즉, 위와 같은 매개변수 지정은 모든 답을 도미라는 결과를 도출하게 만든다.

````python
print(35/49)
````

위와 같이 결과값이 전체 데이터 갯수를 도미 갯수로 나누면 같은 결과값을 나온다.  
즉, 도미만 맞춘 것이다. (빙어를 도미라고 예측한 것)
